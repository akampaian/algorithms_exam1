```python

menu = {
    'beans': 5,
    'meat': 8,
    'Gnuts': 4,

}


def calculate_total_cost(order, memo={}):
    if not order:
        return 0

    order_key = tuple(order)


    if order_key in memo:
        return memo[order_key]

    total_cost = 0
    for item in order:
        if item in menu:
            total_cost += menu[item]


    memo[order_key] = total_cost
    return total_cost


customer_order_1 = ['beans', 'meat', 'Gnuts']
customer_order_2 = ['beans', 'beans', 'Gnuts', 'meat']

total_cost_1 = calculate_total_cost(customer_order_1)
total_cost_2 = calculate_total_cost(customer_order_2)

print(f"Total cost for order 1: ${total_cost_1}")
print(f"Total cost for order 2: ${total_cost_2}")
print(f"Total co")
```

```python
import timeit
import matplotlib.pyplot as plt
import random

# Bubble Sort implementation
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Selection Sort implementation
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]


def generate_input_list(size):
    return [random.randint(1, 1000) for _ in range(size)]


def measure_sorting_time(sort_func, input_list):
    start_time = timeit.default_timer()
    sort_func(input_list)
    end_time = timeit.default_timer()
    return end_time - start_time

# Plotting time complexities against input size
input_sizes = [100, 200, 300, 400, 500]
bubble_sort_times = []
selection_sort_times = []

for size in input_sizes:
    input_list = generate_input_list(size)

    bubble_time = measure_sorting_time(bubble_sort, input_list.copy())
    bubble_sort_times.append(bubble_time)

    selection_time = measure_sorting_time(selection_sort, input_list.copy())
    selection_sort_times.append(selection_time)

plt.plot(input_sizes, bubble_sort_times, label='Bubble Sort')
plt.plot(input_sizes, selection_sort_times, label='Selection Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity Analysis of Sorting Algorithms')
plt.legend()
plt.show()
```

By analyzing the plotted graphs, you can draw conclusions about the time complexity of each algorithm. Typically, Bubble Sort has an average-case and worst-case time complexity of O(n^2), while Selection Sort also has an average-case and worst-case time complexity of O(n^2).

Observations:

Both algorithms have quadratic time complexities, which means their execution time increases quadratically as the input size (n) grows.
As the input size increases, the execution time for both algorithms will grow significantly. However, their performance might differ slightly for various input sizes.
Bubble Sort and Selection Sort are generally not efficient for large input sizes compared to more efficient sorting algorithms like Merge Sort or Quick Sort, which have better time complexities (O(n log n)) in most cases.
Conclusion:
For larger input sizes, both Bubble Sort and Selection Sort might become increasingly inefficient compared to more optimized sorting algorithms due to their quadratic time complexities. Consider using more efficient algorithms for larger datasets.
